```python
import re

def check_password_strength(password):
    score = 0
    feedback = []

    # 1. Length check
    length = len(password)
    if length >= 12:
        score += 2
    elif length >= 8:
        score += 1
        feedback.append("💡 Use 12+ characters for a stronger password.")
    else:
        feedback.append("❌ Too short! Use at least 8 characters.")

    # 2. Uppercase letters
    if re.search(r'[A-Z]', password):
        score += 1
    else:
        feedback.append("❌ Add at least one uppercase letter (A-Z).")

    # 3. Lowercase letters
    if re.search(r'[a-z]', password):
        score += 1
    else:
        feedback.append("❌ Add at least one lowercase letter (a-z).")

    # 4. Digits
    if re.search(r'[0-9]', password):
        score += 1
    else:
        feedback.append("❌ Add at least one number (0-9).")

    # 5. Special characters
    if re.search(r'[!@#$%^&*(),.?\":{}|<>_\-]', password):
        score += 2
    else:
        feedback.append("❌ Add special characters (e.g. @, #, $, !).")

    # 6. Penalize common patterns
    common_passwords = ["password", "123456", "qwerty", "abc123", "letmein", "admin", "welcome"]
    if password.lower() in common_passwords:
        score = 0
        feedback = ["🚫 This is a very common password. Please choose something unique."]

    # Determine strength level
    if score <= 2:
        strength = "🔴 WEAK"
    elif score <= 4:
        strength = "🟡 OKAY"
    else:
        strength = "🟢 STRONG"

    return strength, score, feedback


def display_result(password, strength, score, feedback):
    print("\n" + "="*45)
    print(f"  Password  : {'*' * len(password)}")
    print(f"  Length    : {len(password)} characters")
    print(f"  Score     : {score}/7")
    print(f"  Strength  : {strength}")
    print("="*45)
    if feedback:
        print("  Suggestions to improve:")
        for tip in feedback:
            print(f"    {tip}")
    else:
        print("  ✅ Excellent! Your password is very strong.")
    print("="*45)


def main():
    print("\n╔══════════════════════════════════════════╗")
    print("║     🔐 PASSWORD STRENGTH CHECKER 🔐      ║")
    print("║      Hunar Intern | Cyber Security        ║")
    print("╚══════════════════════════════════════════╝")

    while True:
        print("\nOptions:")
        print("  [1] Check a password")
        print("  [2] Exit")
        choice = input("\nEnter your choice (1 or 2): ").strip()

        if choice == "1":
            password = input("\nEnter the password to check: ")
            if not password:
                print("⚠️  No password entered. Please try again.")
                continue
            strength, score, feedback = check_password_strength(password)
            display_result(password, strength, score, feedback)

        elif choice == "2":
            print("\n👋 Thank you for using Password Strength Checker!")
            print("   Stay secure! — Hunar Intern\n")
            break

        else:
            print("⚠️  Invalid choice. Please enter 1 or 2.")


if __name__ == "__main__":
    main()
```
